```python
from sklearn import datasets
from sklearn.model_selection import train_test_split
from yellowbrick.classifier import ConfusionMatrix
# Importando Tensorflow (pip install tensorflow)
# Importando Keras (pip install keras)
from keras.models import Sequential
from keras.layers import Dense
from keras.utils import np_utils
import numpy as np
from sklearn.metrics import confusion_matrix
```

```python
base = datasets.load_iris()
```

```python
base
```

```python
previsores = base.data
classe = base.target
```

```python
previsores
```

```python
classe
```

```python
classe_dummy = np_utils.to_categorical(classe)
```

```python
classe_dummy
```

```python
X_treinamento, X_teste, y_treinamento, y_teste = train_test_split(previsores, classe_dummy, test_size = 0.3,random_state=0)
```

```python
# Construindo a rede neural
modelo = Sequential()

#adicionando as camadas
modelo.add(Dense(units=5,input_dim=4)) # 5 neuronios
modelo.add(Dense(units=4)) # 4 neuronios
modelo.add(Dense(units=3,activation='softmax')) # 3 neuronios
```

```python
modelo.summary()
```

```python
from keras.utils.vis_utils import plot_model
```

```python
modelo.compile(optimizer='adam',loss='categorical_crossentropy',metrics=['accuracy'])
modelo.fit(X_treinamento, y_treinamento, epochs=1000, validation_data=(X_teste,y_teste))
```

```python
previsoes=modelo.predict(X_teste)
```

```python
previsoes
```

```python
previsoes=(previsoes>0.5)
```

```python
previsoes
```

```python
#v = ConfusionMatrix()
y_teste_matrix=[np.argmax(t) for t in y_teste]
```

```python
y_teste_matrix
```

```python
y_previsao_matrix=[np.argmax(t) for t in previsoes]
```

```python
y_previsao_matrix
```

```python
confusao=confusion_matrix(y_teste_matrix, y_previsao_matrix)
```
