```python
# An√°lise de dados do mercado financeiro

# importando bibliotecas
from pandas_datareader import data,wb
import pandas as pd
import numpy as np
import datetime
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
inicio = datetime.datetime(2006, 1,1)
fim = datetime.datetime(2016, 1 ,1)

# Bank of America
BAC = data.DataReader("BAC","yahoo", inicio, fim)
```

```python
BAC.head()
```

```python
BAC.plot()
```

```python
# CitiGroup
C=data.DataReader("C","yahoo", inicio, fim)

# Goldman Sachs
GS=data.DataReader("GS","yahoo", inicio, fim)

# JPMorgan Chave
JPM=data.DataReader("JPM","yahoo", inicio, fim)

# Morgan StanLey
MS=data.DataReader("MS","yahoo", inicio, fim)

# Wells Fargo
WFC=data.DataReader("WFC","yahoo", inicio, fim)
```

```python
# Criando uma lista dos s√≠mbolos dos tickers (como strings) em ordem alfab√©tica.
tickers=['BAC','C','GS','JPM','MS','WFC']
```

```python
# Concatenando os DataFrames do banco juntos em um √∫nico chamado bank_stocks.
bank_stocks=pd.concat([BAC,C,GS,JPM,MS,WFC],axis=1,keys=tickers)
```

```python
bank_stocks.head()
```

```python
# Definindo os niveis dos nomes das colunas
bank_stocks.columns.names=['Bank Ticker','Stock Info']
```

```python
bank_stocks.head()
```

## An√°lise Explorat√≥ria de Dados

```python
# Qual √© o pre√ßo maximo de fechamento para o estoque de cada banco durante todo o per√≠odo?
bank_stocks.xs(key='Close',axis=1,level='Stock Info').max()
```

```python
retornos = pd.DataFrame()
```

```python
for tick in tickers:
    retornos[tick + 'Return'] = bank_stocks[tick]['Close'].pct_change()
```

```python
retornos.head()
```

```python
# Criando um pairplot
sns.pairplot(retornos[1:])
```

```python
# Quais datas cada a√ß√£o dos bancos teve o melhor e o pior dia de retorno?
retornos.idxmin()
```

## Explorando Visualiza√ß√µes dos Dados

```python
# Criando um gr√°fico de linha mostrando o pre√ßo de fechamento para cada banco para todoo √≠ndice de tempo
sns.set_style('whitegrid')
```

```python
for tick in tickers:
    bank_stocks[tick]['Close'].plot()
plt.legend()
```

```python
# Melhorando o gr√°fico
for tick in tickers:
    bank_stocks[tick]['Close'].plot(figsize=(12,4),label=tick)
plt.legend()
```

```python
# Utilizando o m√©todo xs
bank_stocks.xs(key='Close',level='Stock Info', axis=1).plot(figsize=(12,4))
```

## Verificando as m√©dias m√≥veis

```python
# Tra√ßando a m√©dia de 30 dias para o pre√ßo pr√≥ximo do Bank Of America para o ano de 2008
BAC2008 = BAC['Close'][BAC.index.year==2008]
```

```python
BAC2008_media_movel = BAC2008.rolling(window=30).mean().plot(figsize=(12,6),label='M√©dia m√≥vel de 30 dias')
BAC2008.plot(label='Fechamento')
plt.legend()
```

```python
# Criando um mapa de calor da correla√ß√£o entre os pre√ßos de fechamento das a√ß√µes
matriz_de_correlacao=bank_stocks.xs(key='Close',level='Stock Info',axis=1).corr()
```

```python
sns.heatmap(matriz_de_correlacao,annot=True)
```

```python
# Criando um clustermap para agrupar as correla√ß√µes
sns.clustermap(matriz_de_correlacao, annot=True)
```
