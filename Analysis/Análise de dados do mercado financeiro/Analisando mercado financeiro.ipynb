```python
# Análise de dados do mercado financeiro

# importando bibliotecas
from pandas_datareader import data,wb
import pandas as pd
import numpy as np
import datetime
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
inicio = datetime.datetime(2006, 1,1)
fim = datetime.datetime(2016, 1 ,1)

# Bank of America
BAC = data.DataReader("BAC","yahoo", inicio, fim)
```

```python
BAC.head()
```

```python
BAC.plot()
```

```python
# CitiGroup
C=data.DataReader("C","yahoo", inicio, fim)

# Goldman Sachs
GS=data.DataReader("GS","yahoo", inicio, fim)

# JPMorgan Chave
JPM=data.DataReader("JPM","yahoo", inicio, fim)

# Morgan StanLey
MS=data.DataReader("MS","yahoo", inicio, fim)

# Wells Fargo
WFC=data.DataReader("WFC","yahoo", inicio, fim)
```

```python
# Criando uma lista dos símbolos dos tickers (como strings) em ordem alfabética.
tickers=['BAC','C','GS','JPM','MS','WFC']
```

```python
# Concatenando os DataFrames do banco juntos em um único chamado bank_stocks.
bank_stocks=pd.concat([BAC,C,GS,JPM,MS,WFC],axis=1,keys=tickers)
```

```python
bank_stocks.head()
```

```python
# Definindo os niveis dos nomes das colunas
bank_stocks.columns.names=['Bank Ticker','Stock Info']
```

```python
bank_stocks.head()
```

## Análise Exploratória de Dados

```python
# Qual é o preço maximo de fechamento para o estoque de cada banco durante todo o período?
bank_stocks.xs(key='Close',axis=1,level='Stock Info').max()
```

```python
retornos = pd.DataFrame()
```

```python
for tick in tickers:
    retornos[tick + 'Return'] = bank_stocks[tick]['Close'].pct_change()
```

```python
retornos.head()
```

```python
# Criando um pairplot
sns.pairplot(retornos[1:])
```

```python
# Quais datas cada ação dos bancos teve o melhor e o pior dia de retorno?
retornos.idxmin()
```

```python
# 4 bancos compartilham o mesmo dia para a pior queda, o que aconteceu
# Utilizando o santo google descobrimos o seguinte: https://www.onthisday.com/date/2009/january/20 (eleição do OBAMA)
# Provavelmente alguns investidores tiveram receio
```

```python
# Verificando o maior retorno
retornos.idxmax()
```

```python
# O pior retorno do Bank of America
retornos['BACReturn'].min()
```

```python
# o melhor retorno do Bank of America
retornos['BACReturn'].max() #argmax também pode ser usado
```

```python
# Qual ação podemos classificar como a mais arriscada durante todo o periodo de tempo? Qual podemos classificar como
# a mais arriscada para o ano de 2015?

# Calculando o desvio padrão, normalmente utilizada no mercado financeiro para prever o risco
retornos.std()
```

```python
# O citi group é quem tem o maior risco
```

```python
retornos[(retornos.index.date >= datetime.date(2015,1,1)) & (retornos.index.date < datetime.date(2016,1,1))].std()
```

```python
# Criando um distplot dos restornos de 2015 para o Morgan StanLey
```

```python
sns.distplot(retornos[(retornos.index.date >= datetime.date(2015,1,1)) & (retornos.index.date < datetime.date(2016,1,1))]['MSReturn'])
```

```python
# Criando um distplot dos retornos de 2008 para o CitiGroup
sns.distplot(retornos[(retornos.index.date >= datetime.date(2008,1,1)) & (retornos.index.date < datetime.date(2009,1,1))]['CReturn'])
```

## Explorando Visualizações dos Dados

```python
# Criando um gráfico de linha mostrando o preço de fechamento para cada banco para todoo índice de tempo
sns.set_style('whitegrid')
```

```python
for tick in tickers:
    bank_stocks[tick]['Close'].plot()
plt.legend()
```

```python
# Melhorando o gráfico
for tick in tickers:
    bank_stocks[tick]['Close'].plot(figsize=(12,4),label=tick)
plt.legend()
```

```python
# Utilizando o método xs
bank_stocks.xs(key='Close',level='Stock Info', axis=1).plot(figsize=(12,4))
```

## Verificando as médias móveis

```python
# Traçando a média de 30 dias para o preço próximo do Bank Of America para o ano de 2008
BAC2008 = BAC['Close'][BAC.index.year==2008]
```

```python
BAC2008_media_movel = BAC2008.rolling(window=30).mean().plot(figsize=(12,6),label='Média móvel de 30 dias')
BAC2008.plot(label='Fechamento')
plt.legend()
```

```python
# Criando um mapa de calor da correlação entre os preços de fechamento das ações
matriz_de_correlacao=bank_stocks.xs(key='Close',level='Stock Info',axis=1).corr()
```

```python
sns.heatmap(matriz_de_correlacao,annot=True)
```

```python
# Criando um clustermap para agrupar as correlações
sns.clustermap(matriz_de_correlacao, annot=True)
```
