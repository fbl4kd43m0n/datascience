```python
# Qual o curso com a mensalidade mais alta?
df[df['mensalidade'] == df['mensalidade'].max()]
```

```python
# Quais o curso mais barato do Brasil?
df[df['mensalidade'] == df['mensalidade'].min()]
```

```python
df[df['mensalidade'] == df['mensalidade'].min()][['universidade_nome','uf_busca','cidade_busca','mensalidade']]
```

```python
# Qual a mensalidade mais cara por UF?
df.groupby('uf_busca')['mensalidade'].max()
```

```python
# Qual a mensalidade mais barata por UF?
df.groupby('uf_busca')['mensalidade'].min()
```

```python
# Mostrando todos os dados dos cursos mais caros e baratos de cada UF
df.loc[df.groupby('uf_busca')['mensalidade'].idxmax]
```

```python
df.loc[df.groupby('uf_busca')['mensalidade'].idxmin]
```

```python
# Mensalidades entre 100 e 300
df[(df['mensalidade'] >= 100) & (df['mensalidade'] <=300)]
```

```python
# Mensalidades entre 100 e 200
df[(df['mensalidade'] >= 100) & (df['mensalidade'] <=200)]
```

```python
# Pegando as 5 maiores mensalidade utilizando outra técnica
df.nlargest(5, 'mensalidade')
```

```python
# Pegando as 5 menores mensalidade utilizando outra técnica
df.nsmallest(5, 'mensalidade')
```
