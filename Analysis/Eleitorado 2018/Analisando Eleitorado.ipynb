
# Análise do Eleitorado Brasileiro

```python
# fonte: http://www.tse.jus.br/eleicoes/estatisticas/estatisticas-eleitorais
import pandas as pd
```

```python
eleitorado=pd.read_csv('eleitorado_municipio_2018.csv', sep=';', encoding='latin1')
```

```python
eleitorado.head(3).T
```

```python
eleitorado.info()
```

```python
# As 5 cidades com maior quantidade de eleitores
eleitorado.nlargest(5,'QTD_ELEITORES')
```

```python
print('Eleitoras....:', eleitorado['QTD_ELEITORES_FEMININO'].sum())
print('Eleitores....:', eleitorado['QTD_ELEITORES_MASCULINO'].sum())
print('Não informado:', eleitorado['QTD_ELEITORES_NAOINFORMADO'].sum())
```

```python
# Criar variaveis para calcular o percentual
tot_eleitores = eleitorado['QTD_ELEITORES'].sum()
tot_fem = eleitorado['QTD_ELEITORES_FEMININO'].sum()
tot_masc = eleitorado['QTD_ELEITORES_MASCULINO'].sum()
tot_nao = eleitorado['QTD_ELEITORES_NAOINFORMADO'].sum()
```

```python
print('Eleitoras....:', ((tot_fem/tot_eleitores)*100).round(2))
print('Eleitores....:', ((tot_masc/tot_eleitores) *100).round(2))
print('Não informado:', ((tot_nao/tot_eleitores)*100).round(2))
```

```python
# Quantos municípios com mais homens que mulheres?
eleitorado[eleitorado['QTD_ELEITORES_MASCULINO'] > eleitorado['QTD_ELEITORES_FEMININO']].count()
```

```python
# Criando uma coluna nova para indicar a relação fem/masc
eleitorado['RELACAO_FM'] = eleitorado['QTD_ELEITORES_FEMININO'] / eleitorado['QTD_ELEITORES_MASCULINO']
```

```python
eleitorado['RELACAO_FM'].head()
```

```python
# Quais são os municípios com maior relação f/m
eleitorado.nlargest(5,'RELACAO_FM').T
```

```python
# Quais os municipios com a menor relação f/m
eleitorado.nsmallest(5,'RELACAO_FM').T
```

```python
# Criando novo dataframe só com municípios do Brasil
df_brasil=eleitorado[eleitorado['CD_PAIS']==1].copy()
```

```python
df_brasil.head(3).T
```

```python
# Quais são os municípios com maior relação f/m
df_brasil.nlargest(5,'RELACAO_FM').T
```

```python
# Quais são os municípios com maior relação f/m
df_brasil.nsmallest(5,'RELACAO_FM').T
```

```python
# plotar gráfico de distribuição da relação f/m
%matplotlib inline
df_brasil['RELACAO_FM'].plot.hist()
```

```python
df_brasil['RELACAO_FM'].plot.hist(bins=100)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# plotar a distruição da relação f/m
sns.distplot(df_brasil['RELACAO_FM'],bins=100,color='red')
#kde=False se quiser tirar a curva de aproximação
plt.title('Relação Eleitoras/Eleitores',fontsize=18)
plt.xlabel('Eleitoras/Eleitores', fontsize=14)
plt.ylabel('Frequência',fontsize=14)
plt.axvline(1.0,color='black',linestyle='--')
```

```python
# verificar a distribuição f/m mas mostrando os pontos (municipios)
sns.swarmplot(data=df_brasil,x='NM_REGIAO',y='RELACAO_FM')
plt.axhline(1.0, color='black', linestyle='--')
```
