# Análise dos dados de reembolso dos senadores - 2019

```python
import pandas as pd
```

```python
df=pd.read_csv('2019.csv', sep=';', encoding='latin1', skiprows=1, decimal=',')
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
# Qual o total de reembolso
df['VALOR_REEMBOLSADO'].sum()
```

```python
# Quantos reembolsos foram solicitados por cada senador?
df['SENADOR'].value_counts()
```

```python
# Os top 5 senadores que mais pediram reembolso
df['SENADOR'].value_counts().head(5)
```

```python
# Quanto cada senador solicitou de reembolso em 2019?
df.groupby('SENADOR')['VALOR_REEMBOLSADO'].sum().sort_values(ascending=False)
```

```python
# Os top 5 senadores com maiores valores de reembolso
#df.groupby('SENADOR')['VALOR_REEMBOLSADO'].sum().sort_values(ascending=False).head()
df.nlargest(5, 'VALOR_REEMBOLSADO').T
```

```python
# Os 5 menores reembolsos
df.nsmallest(5, 'VALOR_REEMBOLSADO').T
```

```python
# Quantidade de reembolsos por tipos de despesa (TIPO_DESPESA)
df['TIPO_DESPESA'].value_counts()
```

```python
# Quanto foi reembolsado por cada tipo de despesa?
df.groupby('TIPO_DESPESA')['VALOR_REEMBOLSADO'].sum().sort_values(ascending=False)
```
